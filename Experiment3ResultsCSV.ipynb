```python
from sklearn.datasets import load_iris
import RandomForest as RF
import os.path
import os
from datetime import datetime
import bz2
import _pickle as cPickle
import numpy as np
import random
import csv

today = datetime.now()

dataset = "unsw_nb15" # iris kdd  nsl_kdd  unsw_nb15 cic_ids_2017
problem = "AD" # AD AC

if dataset == "kdd" and problem == "AD" :
    # num_trees = 65
    criterion_input = 'entropy' #["gini", "entropy"]
    ensemble_method = "SV"

if dataset == "kdd" and problem == "AC" :
    # num_trees = 53
    criterion_input = 'entropy' #["gini", "entropy"]
    ensemble_method = "WV"

if dataset == "nsl_kdd" and problem == "AD" :
    # num_trees = 65
    criterion_input = 'gini' #["gini", "entropy"]
    ensemble_method = "SV"

if dataset == "nsl_kdd" and problem == "AC" :
    # num_trees = 31
    criterion_input = 'entropy' #["gini", "entropy"]
    ensemble_method = "SV"

if dataset == "unsw_nb15" and problem == "AD" :
    # num_trees = 93
    criterion_input = 'entropy' #["gini", "entropy"]
    ensemble_method = "WV"

if dataset == "unsw_nb15" and problem == "AC" :
    # num_trees = 61
    criterion_input = 'entropy' #["gini", "entropy"]
    ensemble_method = "SV"

if dataset == "cic_ids_2017" and problem == "AD" :
    criterion_input = 'entropy' #["gini", "entropy"]
    ensemble_method = "WV"
    # num_trees = 77

if dataset == "cic_ids_2017" and problem == "AC" :
    criterion_input = 'entropy' #["gini", "entropy"]
    ensemble_method = "WV"
    # num_trees = 95

methods = [ "RF_S_DTs_A_All", "RF_S_DTs_WA_All", "RF_S_DTs_A", "RF_S_DTs_WA"]


numTrees = []
results_3_acc = []
results_3_prec = []
results_3_recall = []
results_3_f1 = []

testingAcc = []
prec = []
recall = []
f1 = []
my_os = "mac" #win mac

separator = "\\"

if my_os == "mac":
    separator = "//"

folderName = 'Experiments 3'  + separator + dataset + separator

no_runs = 5

for i in range(len(methods)):

    testingAcc.append([])
    prec.append([])
    recall.append([])
    f1.append([])
    numTrees.append([])
    results_3_acc.append([])
    results_3_prec.append([])
    results_3_recall.append([])
    results_3_f1.append([])


    for k in range(1,no_runs+1):
        print(i,"   ",k)

        file_name = folderName + "Experiment_3_RF_"+dataset+"_"+problem+"_"+criterion_input+"_"+ensemble_method+"_"+methods[i]+"_"+str(k)+".pbz2"

        #f = open(file_name, "rb")
        f = bz2.BZ2File(file_name, "rb")
        algorithms = cPickle.load(f)
        f.close()
        if len(algorithms)==7:
            algorithms1 = []
            for l in range(len(algorithms)):
                temp = algorithms[l]
                for z in range(len(temp)):
                    algorithms1.append(temp[z])
            algorithms = algorithms1

        for h in range(len(algorithms)):
            if k == 1:
                numTrees[i].append(algorithms[h].size)
                testingAcc[i].append(algorithms[h].statisticsRF.accuracy)
                prec[i].append(algorithms[h].statisticsRF.precision*100)
                recall[i].append(algorithms[h].statisticsRF.recall*100)
                f1[i].append(algorithms[h].statisticsRF.f1*100)
            else:
                testingAcc[i][h] += algorithms[h].statisticsRF.accuracy
                prec[i][h] += algorithms[h].statisticsRF.precision*100
                recall[i][h] += algorithms[h].statisticsRF.recall*100
                f1[i][h] += algorithms[h].statisticsRF.f1*100
            # if algorithms[h].statisticsRF.accuracy > max:
            #     max = algorithms[h].statisticsRF.accuracy
            #     maxNumTrees = algorithms[h].size

        #print(methods[i]," ", max , " ", maxNumTrees)
    for h in range(len(algorithms)):
        results_3_acc[i].append(round(testingAcc[i][h]/no_runs, 3))
        results_3_prec[i].append(round(prec[i][h]/no_runs, 3))
        results_3_recall[i].append(round(recall[i][h]/no_runs, 3))
        results_3_f1[i].append(round(f1[i][h]/no_runs, 3))


f = open('Experiments 3'  + separator + dataset+"_"+problem+"_results_3.csv", 'w')
writer = csv.writer(f)

for k in range(len(results_3_acc)):
    writer.writerow(numTrees[k])
    writer.writerow(results_3_acc[k])
    writer.writerow(results_3_prec[k])
    writer.writerow(results_3_recall[k])
    writer.writerow(results_3_f1[k])

#writer.writerow(avg_2_1)
f.close()
```
